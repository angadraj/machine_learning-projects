```python
cfeat=['purpose']
```

```python
final_df=pd.get_dummies(df,columns=cfeat,drop_first=True)
```

```python
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop('not.fully.paid',axis=1)
y=final_df['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
dpred=dtc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(dpred,y_test))
print(confusion_matrix(dpred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
rpred=rfc.predict(X_test)
```

```python
print(classification_report(rpred,y_test))
print(confusion_matrix(rpred,y_test))
```
